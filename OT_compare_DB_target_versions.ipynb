## Merge and check diffs
* Only look at cases with match in both or in newer?

```python
df_old.columns
```

```python
KeepColsList = ['id', 'children',  'dbXRefs', #'code',
                'name',
       'obsoleteTerms', 'parents','has_children']
df = pd.merge(df_old[KeepColsList],df_new[KeepColsList],on="id",how="inner",suffixes = ('_old', '_new'),)
print(df["id"].nunique())
df
```

```python
# df.loc[df["has_children_new"]!=df["has_children_old"]]
df["mismatch"] = df["has_children_new"]!=df["has_children_old"]
print(df["mismatch"].agg(["mean","sum","size"]).round(3))
df.loc[df["mismatch"]] ## 396 rows
```

```python
## warning! free text!!!! 
s1_candidate_names = pd.read_csv("outputs/S1-candidate_errors.csv")["name"]
print(len(s1_candidate_names))
s1_candidate_names
```

```python
df.loc[(df.name_new.isin(s1_candidate_names)) | (df.name_old.isin(s1_candidate_names))] # 3 more rows if adding in old names
## 
```

#### For the ~ 600 matched (candidate novel) dissases (partial match warning), we see ~50% higher rate of overrepresentation in mismatches!
i.e 4.6% mismatch vs 3% overall. 0.5 effect size 

```python
df.loc[(df.name_new.isin(s1_candidate_names)) | (df.name_old.isin(s1_candidate_names))]["mismatch"].agg(["mean","sum","size"]).round(3)
```

### data version usedin model (July 2022?)

```python
df_m = pd.read_parquet("./diseases",engine="fastparquet") # read raw, not proc
# df_new = pd.read_parquet("db_versions/23.12/diseases",engine="fastparquet")
df_m["has_children"] = df_m["children"].apply(lambda x: len(x)>0).astype(int)
df_m
```

```python
df2 = pd.merge(df_m[KeepColsList],df_new[KeepColsList],on="id",how="inner",suffixes = ('_m', '_new'),)
print(df2["id"].nunique())
df2
```

```python
df2["mismatch"] = df2["has_children_new"]!=df2["has_children_m"]
print(df2["mismatch"].agg(["mean","sum","size"]).round(3))
df2.loc[df2["mismatch"]] ## 242 rows
```

## using our data - effect size is even higher  - 2x!! 

* Compare using Z Score Calculator for 2 Population Proportions (with and without candidates)
    *   https://www.socscistatistics.com/tests/ztest/default2.aspx
  * The value of z is -3.2392. The value of p is .0006. The result is significant at p < .01.

```python
df2["candidate"] = (df2.name_new.isin(s1_candidate_names)) | (df2.name_m.isin(s1_candidate_names))
df2.loc[df2["candidate"] ]["mismatch"].agg(["mean","sum","size"]).round(3)
```

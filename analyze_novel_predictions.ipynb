### analyze and view subsets of novel predicted disease subtypes
* Expects output from : `disease_hierarchy-EvalModels.ipynb`



* `has_children` = target annotation in raw data/OT
* `CV_pred` - our models predictions of subtype (labelled as 1 if 8/8 predicted OOF)
* `CV_mean` - our models prediction score/probability, of disease having subtype

````python
import pandas as pd
````

* df_preds is `df_candidate_errors` from `disease_hierarchy-EvalModels`. 

```df_candidate_errors = df.loc[(df["CV_pred"] != df["has_children"]) & (df["CV_stable"])]```


````python
df_preds = pd.read_csv("./outputs/S1-candidate_errors.csv")#.gz") # candidate predictions
df_preds
````

###### take subset of columns for easier viewing

* `has_children` = target annotation in raw data/OT
* `CV_pred` - our models predictions of subtype (labelled as 1 if 8/8 predicted OOF)
* `CV_mean` - our models prediction score/probability, of disease having subtype

````python
df = df_preds[['name', "source",'has_children', 'CV_pred', 'CV_mean']].copy() 
#,'ancestors_sub_parents',

# df["name_wc"] = df["name"].str.split().str.len()
# df["name_chars"] = df["name"].str.len()
# df["desc_wc"] = df["description"].str.split().str.len()
# df["-_count"] = df["name"].str.split("-").str.len()
df["type_or_syndrome"] = df["name"].str.contains("type|syndrome",case=False,na=False)
df["benign"] = df["name"].str.contains("benign",case=False,na=False)
df["abnormal_anomal"] = df["name"].str.contains("abnormal|anomal",case=False,na=False)

df
````

````python
df.select_dtypes(["number","bool"]).agg(["mean","max"]).round(4)
````

````python
df.groupby(["source"])[["has_children","CV_pred"]].agg(["mean","size"]).round(2).sort_values(('has_children', 'size'),ascending=False) # ,"sum"
````

````python
df[df["CV_pred"]>0].head(15)
````

````python
df[(df["type_or_syndrome"]) & (df["CV_pred"]>0)].head(15)
````

````python
print("cases with type/syndrome with either prediction")
df[df["type_or_syndrome"]]#.head(15)
````

````python
print("benign cases predicted to not be subtypes")
df[(df["benign"]) & (df["CV_pred"]==0)].tail(8)
````
